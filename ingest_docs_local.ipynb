```python
import os, sys, subprocess, json
from pathlib import Path
from datetime import datetime
from typing import List, Dict, Tuple
import pdfplumber, io, re, numpy as np
from openai import OpenAI

def execute_sql(query: str, fetch: bool = False):
    cmd = ["docker-compose", "-f", "docker-compose.local.yml", "exec",
           "postgres", "psql", "-U", "llm_user", "-d", "system_llm", "-t", "-c", query]
    result = subprocess.run(cmd, capture_output=True, text=True, encoding="utf-8", errors="replace", cwd=".")
    if result.returncode != 0:
        raise Exception(f"SQL Error: {result.stderr}")
    return result.stdout.strip() if fetch else result.stdout.strip()

print("✅ Setup complete")
```
